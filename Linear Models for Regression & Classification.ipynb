## Simple linear regression using ordinary least square(OLS)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Target/dependent variable is denoted by Y
#Feature & Target is denoted by equation (Y = w0 + w1X1+ w2X2+ ...)
# w0-->intercept and w's--> coef. of each variable
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Generating Regession Dataset
```

```python
from sklearn.datasets import make_regression
```

```python
X,Y = make_regression(n_features=1, noise=10, n_samples=1000)
```

```python
plt.xlabel=('feature-X')
plt.ylabel=('target-Y')
plt.scatter(X,Y,s=5)
```

```python
lr = LinearRegression()
```

```python
#Train Model
```

```python
lr.fit(X,Y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
#Predicting using trained model
```

```python
pred = lr.predict(X)
```

```python
plt.scatter(X,Y,s=5,label='training')
plt.scatter(X,pred,s=5,label='Prediction')
plt.xlabel('feature - X')
plt.ylabel('Target - Y')
plt.legend()
plt.show()
```
