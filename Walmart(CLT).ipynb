```python
Avgamt_Male = AvgAmt_Gender[AvgAmt_Gender['Gender']=='M']
Avgamt_Female = AvgAmt_Gender[AvgAmt_Gender['Gender']=='F']
```

```python
#finding the sample (sample size=1000) for avg amount purchase for males and females

Gender=['M','F']
sample_size = 1000

num_repitations = 1000
male_means = []
female_means = []

for i in range(num_repitations):
    male_mean = Avgamt_Male.sample(sample_size, replace=True)['Purchase'].mean()
    female_mean = Avgamt_Female.sample(sample_size, replace=True)['Purchase'].mean()
    
    male_means.append(male_mean)
    female_means.append(female_mean)
```

```python
male_means
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,6))

axs[0].hist(male_means, bins=35)
axs[1].hist(female_means, bins=35)
axs[0].set_title('Male distribution of means, sample size 1000')
axs[1].set_title('Female distribution of means, sample size 1000')

plt.show()
```

```python
#calculating 90% confidence interval for sample size 1000
```

```python
#Taking z value at 90, 95 and 99% of confidence interval as:
z90 = 1.645 #90% confidence interval
z95 = 1.960 #95% confidence interval
z99 = 2.576 #99% confidence interval

print("Population Avg Purchase for males: {:.2f}".format(Avgamt_Male['Purchase'].mean()))
print("Population Avg Purchase for females: {:.2f}\n".format(Avgamt_Female['Purchase'].mean()))

print("Sample Avg Purchase for male: {:.2f}".format(np.mean(male_means)))
print("Sample Avg Purchase for female: {:.2f}\n".format(np.mean(female_means)))

print("Sample std for male: {:.2f}".format(pd.Series(male_means).std()))
print("Sample std for female: {:.2f}\n".format(pd.Series(female_means).std()))

print("Sample std error for male: {:.2f}".format(pd.Series(male_means).std()/np.sqrt(1000)))
print("Sample std error for female: {:.2f}\n".format(pd.Series(female_means).std()/np.sqrt(1000)))

sample_mean_male = np.mean(male_means)
sample_mean_female = np.mean(female_means)

sample_std_male = pd.Series(male_means).std()
sample_std_female = pd.Series(male_means).std()

sample_std_error_male = pd.Series(male_means).std()/np.sqrt(1000)
sample_std_error_female = pd.Series(female_means).std()/np.sqrt(1000)

upper_limit_male = sample_mean_male + z90*sample_std_error_male
lower_limit_male = sample_mean_male - z90*sample_std_error_male

upper_limit_female = sample_mean_female + z90*sample_std_error_female
lower_limit_female = sample_mean_female - z90*sample_std_error_female

print("Male_Purchace_range: ",[lower_limit_male, upper_limit_male])
print("Female_Purchase_range: ",[lower_limit_female, upper_limit_female])
```
